```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('most_valuable_teams_cleaned.csv', low_memory=False)
```

## Distribución del Valor de Mercado por Competición (Liga)

```python
# Agrupa los datos por competición y suma el valor de mercado
market_value_by_competition = df.groupby('Competition')['Market_value'].sum().sort_values(ascending=False)

# Crea un gráfico de barras para mostrar los valores de mercado por competición
plt.figure(figsize=(10,4))
sns.barplot(x=market_value_by_competition.index, 
            y=market_value_by_competition.values, 
            hue=market_value_by_competition.index, 
            palette="viridis", 
            legend=False)

# Añáde títulos y etiquetas
plt.title('Valor de Mercado Total por Competición', fontsize=16)
plt.xlabel('Competición', fontsize=12)
plt.ylabel('Valor de Mercado (€ millones)', fontsize=12)

# Rota las etiquetas del eje X para mejorar la visualización
plt.xticks(rotation=45)
plt.show()
```

## Relación entre Tamaño del Plantel y Valor de Mercado

```python
# Crea un gráfico de dispersión para analizar la relación entre el valor de mercado y el tamaño del plantel
plt.figure(figsize=(10,6))
sns.scatterplot(x='Squad_size', y='Market_value', data=df, hue='Competition', palette="coolwarm", s=100)

# Añáde título y etiquetas
plt.title('Relación entre Tamaño del Plantel y Valor de Mercado', fontsize=16)
plt.xlabel('Tamaño del Plantel', fontsize=12)
plt.ylabel('Valor de Mercado (€ millones)', fontsize=12)

# Muestra el gráfico
plt.show()
```

## Relación entre Edad Promedio del Plantel y Valor de Mercado

```python
# Crea un gráfico de dispersión para la relación entre la edad promedio y el valor de mercado
plt.figure(figsize=(10,6))
sns.scatterplot(x='Age', y='Market_value', data=df, hue='Competition', palette="Set2", s=100)

# Añáde título y etiquetas
plt.title('Relación entre Edad Promedio del Plantel y Valor de Mercado', fontsize=16)
plt.xlabel('Edad Promedio del Plantel', fontsize=12)
plt.ylabel('Valor de Mercado (€ millones)', fontsize=12)

# Muestra el gráfico
plt.show()
```

## Valor de los 18 Jugadores Más Caros como Porcentaje del Valor Total

```python
# Crea un gráfico de dispersión para mostrar la relación entre el valor total del plantel y el porcentaje de los 18 jugadores más caros
plt.figure(figsize=(10,6))
sns.scatterplot(x='Market_value', y='Share_of_MV', data=df, hue='Competition', palette="Accent", s=100)

# Añáde título y etiquetas
plt.title('Porcentaje del Valor Total en los 18 Jugadores Más Caros', fontsize=16)
plt.xlabel('Valor de Mercado (€ millones)', fontsize=12)
plt.ylabel('Porcentaje del Valor en los 18 Jugadores Más Caros (%)', fontsize=12)

# Muestra el gráfico
plt.show()
```
